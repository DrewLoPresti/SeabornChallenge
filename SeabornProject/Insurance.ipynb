```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
insData = pd.read_csv('Data/insurance.csv')
```

```python
insData.tail()
```

```python
insData.describe()
```

```python
cmatrix = insData.corr()
f, ax = plt.subplots(figsize=(20, 9))
sns.heatmap(cmatrix, vmax=.8, annot=True);
```

```python
sns.set()
cols = ['charges', 'age', 'bmi', 'children']
sns.pairplot(insData[cols])
plt.show();
```

```python
# color-code by 'smoker' : bmi/y #
sns.scatterplot(x=insData['bmi'],
                y=insData['charges'], hue=insData['smoker'])
```

```python
sns.swarmplot(x=insData['smoker'],
              y=insData['charges'])
```

```python
sns.swarmplot(x=insData['children'],
              y=insData['charges'])
```

```python
sns.swarmplot(x=insData['region'],
              y=insData['charges'])
```

```python
sns.pairplot(insData, hue = 'smoker')
plt.show();
```

```python
sns.pairplot(insData, hue = 'region')
plt.show();
```
